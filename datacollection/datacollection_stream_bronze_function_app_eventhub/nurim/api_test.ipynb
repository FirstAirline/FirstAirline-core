```python
import requests
 
# ▶ 포털에서 받은 Encoding된 서비스키 그대로 입력
serviceKey = "wMeQErx7JspmwkeUt7B5Q1k5W81LN4lnBP9u6B5WZjuyvSWtFdXP7B8vnd47MmndQrqfWzI0TmXLKUlIwKtRgg=="
 
# End Point (오퍼레이션 포함)
url = "https://apis.data.go.kr/B551177/StatusOfArrivals/getArrivalsCongestion"
 
params = {
    "serviceKey": serviceKey,
    "terno": "T1",           # T1=제1터미널, T2=제2터미널
    "numOfRows": 10,
    "pageNo": 1,
    "type": "json"           # json이 기본, 지원 안되면 xml
}
 
response = requests.get(url, params=params)
 
print(f"status_code: {response.status_code}")
print(f"Content-Type: {response.headers.get('Content-Type')}")
print("-" * 30)
 
text = response.text.strip()
if not text:
    print("❌ 응답이 비어 있습니다.")
elif "OpenAPI_ServiceResponse" in text:
    print("❌ 에러 응답(XML):")
    print(text)
    # 추가: 에러 코드 출력
    import re
    errcode = re.search(r"<returnReasonCode>(\d+)</returnReasonCode>", text)
    if errcode:
        print(f"→ 에러 코드: {errcode.group(1)}")
    errmsg = re.search(r"<errMsg>(.*?)</errMsg>", text)
    if errmsg:
        print(f"→ 에러 메시지: {errmsg.group(1)}")
else:
    try:
        data = response.json()
        header = data.get('response', {}).get('header', {})
        if header.get('resultCode') == '00':
            items = data['response']['body']['items']
            if not items:
                print("데이터가 없습니다.")
            else:
                for i, item in enumerate(items, 1):
                    print(f"--- 항공편 {i} ---")
                    print(f"공항: {item.get('airport', '')}")
                    print(f"입국장게이트: {item.get('entrygate', '')}")
                    print(f"편명: {item.get('flightid', '')}")
                    print(f"도착예정: {item.get('scheduletime', '')}")
                    print(f"도착실제: {item.get('estimatedtime', '')}")
                    print(f"내국인: {item.get('korean', '')}")
                    print(f"외국인: {item.get('foreigner', '')}")
                    print(f"여객기도착게이트: {item.get('gatenumber', '')}")
                    print(f"터미널: {item.get('terno', '')}")
                    print()
        else:
            print(f"❌ API 오류: {header.get('resultMsg')} ({header.get('resultCode')})")
    except Exception as e:
        print("❌ JSON 파싱 실패. 응답 원문 출력:")
        print(text) 
```
